# 1º Exercício Atividade Prática

```python
# Print de identificação
print('Bem-vindo a loja do Wélyson Filipe de Oliveira')
# Criação das variáveis de entrada de dados
valor_produto = float(input('Entre com o valor do produto: '))
quantidade = int(input('Entre com a quantidade do produto: '))
desc_produto = 0 # Criação de uma variável para armazenar o valor do desconto
# Tratamento das condições
if quantidade < 200:
  desc_produto = 0
elif quantidade >= 200 and quantidade < 1000:
  desc_produto = 0.05
elif quantidade >= 1000 and quantidade < 2000:
  desc_produto = 0.10
else:
  desc_produto = 0.15
# Apresentando os resultados
total_sem_desc = valor_produto * quantidade
print('O valor SEM desconto é de: R${:.2f}'.format(total_sem_desc))
total_com_desc = total_sem_desc - total_sem_desc * desc_produto
print('O valor COM desconto é de: R${:.2f}'.format(total_com_desc))
```

# 3º Exercício da Atividade Prática

```python
# Adicionando a identificação
print('Bem-Vindo ao Petshop do Wélyson Filipe de Oliveira')

# Início da função cachorro_peso()
def cachorro_peso():
  base = 0
  while True:
    try:
      peso = int(input('Entre com o peso do cachorro: '))
      if peso < 3:
        base = base + 40
      elif (peso >= 3) and (peso < 10):
        base = base + 50
      elif (peso >= 10) and (peso < 30):
        base = base + 60
      elif (peso >= 30) and (peso < 50):
        base = base + 70
      elif peso > 50:
        print('Não aceitamos cachorros tão grandes')
        continue #retorna para o início do lço
      else:
        print('Por favor entre com o peso do cachorro novamente')
        continue #retorna para o início do laço
      return peso, base #retorna o peso e a base
    except ValueError:
      print('Pare de digitar valores não inteiros')
# Fim da função cachorro_peso()

# Início da função cachorro_pelo()
def cachorro_pelo():
  while True:
    pelo = input('Entre com o pelo do cachorro \n' +
                'c - Pelo curto \n' +
                'm - Pelo médio \n' +
                'l - Pelo longo \n' +
                '>> ')
    pelo = pelo.lower() # Para todas as letras se tornarem minusculas
    pelo = pelo.strip() # Tira os espaçamentos
    if pelo == 'c':
      return 1
    elif pelo == 'm':
      return 1.5
    elif pelo == 'l':
      return 2
    else:
      print('Pare de digitar opções diferentes de c/m/l')
      continue # Voltar para o inicio do laço
# Fim da função cachorro_pelo()

# Início da função cachorro_extra
def cachorro_extra():
  acumulador = 0
  while True:
    extra = input('Deseja adicionar mais algum serviço? \n' +
                  '1 - Corte de unhas - R$10,00 \n' +
                  '2 - Escovar dentes - R$12,00 \n' +
                  '3 - Limpeza de orelhas - R$15,00 \n' +
                  '0 - Não desejo mais nada \n' +
                  '>> ')
    if extra == '0':
      return acumulador
    elif extra == '1':
      acumulador = acumulador + 10
      continue # Voltar para o inicio do laço
    elif extra == '2':
      acumulador = acumulador + 12
      continue
    elif extra == '3':
      acumulador = acumulador + 15
      continue
    else:
      print('Pare de digitar opções diferentes de 0/1/2/3')
# Fim da função cachorro_extra

# Início do Programa Principal
peso, base = cachorro_peso()
pelo = cachorro_pelo()
extra = cachorro_extra()
total = (base * pelo) + extra
print('Total a pagar (R$): {:.2f} (peso: {} * pelo: {} + adicional(is): {})'.format(total,peso,pelo,extra))
```

```python
# Criando variáveis globais
lista_colaborador = []
id_colaborador = 0
# Fim variáveis globais
def cadastrar_colaborador(id_global):
  print('Bem_vindo ao Menu Cadastrar Colaborador')
  print('ID do colaborador: {}'.format(id_global))
  nome = input('Por favor entre com o nome do colaborador: ')
  setor = input('Por favor entre com o nome do setor: ')
  salario = float(input('Por favor entre com o pagamento (R$): '))
  dic_colaborador = {'id': id_global,
                     'nome': nome,
                     'setor': setor,
                     'pagamento': salario}
  lista_colaborador.append(dic_colaborador.copy())

# Início da função consultar_colaborador()

def consultar_colaborador():
  print('Bem_vindo ao Menu Consultar Colaborador')
  while True:
    opcao_consultar = input('Escolha a opção desejada: \n' +
                              '1 - Consultar TODOS os Colaborador(es) \n' +
                              '2 - Consultar Colaborador por ID \n' +
                              '3 - Consultar Colaborador(es) por SETOR \n' +
                              '4 - Retornar \n' +
                              '>> ')
    if opcao_consultar == '1':
      print('Você escolheu consultar TODOS os colaboradores')
      for colaborador in lista_colaborador: # Irá varrer toda a lista de colaborador
        print('-'*30)
        for chave, valor in colaborador.items(): # Irá varrer todos os conjuntos de chave e valor do dicionario colaborador
          print('{}: {}'.format(chave,valor))
        print('-'*30)

    elif opcao_consultar == '2':
      print('Você escolheu a opção consultar Colaborador por ID')
      valor_id = int(input('Entre com o ID desejado: '))
      for colaborador in lista_colaborador:
        if colaborador['id'] == valor_id:
          print('-'*30)
          for chave, valor in colaborador.items(): # Irá varrer todos os conjuntos de chave e valor do dicionario colaborador
            print('{}: {}'.format(chave,valor))
        print('-'*30)

    elif opcao_consultar == '3':
      print('Você escolheu a opção consultar Colaborador por SETOR')
      valor_id = (input('Entre com o SETOR desejado: '))
      for colaborador in lista_colaborador:
        if colaborador['setor'] == valor_id:
          print('-'*30)
          for chave, valor in colaborador.items(): # Irá varrer todos os conjuntos de chave e valor do dicionario colaborador
            print('{}: {}'.format(chave,valor))
        print('-'*30)

    elif opcao_consultar == '4':
        return # Sai da função consultar_produto() e volta para o programa principal
    else:
        print('Opção Inválida. Tente Novamente')
        continue # Volta para o início do laço

# Fim da função consultar_colaborador()

# Início da função remover_colaborador()
def remover_colaborador():
  print('Bem-vindo ao Menu Remover Colaborador')
  valor_id = int(input('Digite o ID do colaborar que deseja remover: '))
  for colaborador in lista_colaborador:
    if colaborador['id'] == valor_id:
      lista_colaborador.remove(colaborador)
      print('Colaborador removido com sucesso!')
# Fim da função remover_colaborador()

# Início do Programa Principal
print('Bem-vindo ao Controle de Colaboradores do Wélyson Filipe de Oliveira')
while True:
  opcao_colaborador = input('Escolha a opção desejada: \n' +
                            '1 - Cadastrar Colaborador \n' +
                            '2 - Consultar Colaborador(es) \n' +
                            '3 - Remover Colaborador \n' +
                            '4 - Sair \n' +
                            '>> ')
  if opcao_colaborador == '1':
    id_colaborador = id_colaborador + 1
    cadastrar_colaborador(id_colaborador)
  elif opcao_colaborador == '2':
    consultar_colaborador()
  elif opcao_colaborador == '3':
    remover_colaborador()
  elif opcao_colaborador == '4':
    break # Encerra o laço principal e o programa acaba
  else:
    print('Opção Inválida. Tente Novamente')
    continue # Retorna para o começo do laço
# Fim do Programa Principal
```
